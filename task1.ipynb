```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
df = pd.read_excel("normalized_uber_data_enhanced (1).xlsx")
```

```python
hour_data = df.groupby("Hour").size()

plt.figure()
hour_data.plot(kind="bar", color="skyblue")

plt.title("Ride Demand by Hour")
plt.xlabel("Hour (0–23)")
plt.ylabel("Number of Rides")

plt.show()
```

```python
slot_data = df["Time_Slot"].value_counts()

plt.figure()
slot_data.plot(kind="bar", color=["orange", "green", "blue", "red"])

plt.title("Ride Distribution by Time Slot")
plt.xlabel("Time Slot")
plt.ylabel("Number of Rides")

plt.show()
```

GRAPH 1: Ride Demand by Hour

🔹 Insight:Ride demand is not uniform throughout the day and shows clear variation across different hours.
The graph indicates that peak demand occurs during specific hours, typically in the morning and evening, which align with daily commuting times.
During late-night and early morning hours, the number of rides is significantly lower, indicating reduced activity.

🔹 Conclusion:This shows that Uber usage is highest during commuting hours, and resources like drivers should be allocated more during these peak periods.

GRAPH 2: Ride Distribution by Time Slot

🔹 Insight:The data is grouped into broader categories such as morning, afternoon, evening, and night, making trends easier to understand.
Certain time slots (especially afternoon and evening) show higher ride counts compared to others.
Night time has comparatively lower demand, indicating reduced ride activity during those hours.

🔹 Conclusion:This graph confirms that ride demand is concentrated in specific time slots, helping identify high-demand periods for better operational planning.
