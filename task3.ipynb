INSIGHT

🔹 1. Booking by Vehicle Type
Auto and Go Mini have the highest number of bookings, indicating they are the most preferred and affordable options.
Premium vehicles like Premier Sedan and Uber XL have lower bookings, suggesting fewer users opt for higher-cost rides.

Insight:
Users prefer cost-effective ride options over luxury services.

🔹 2. Rides by Time Slot
Afternoon has the highest ride demand, followed by morning and evening.
Night has the lowest demand, showing reduced activity during late hours.

Insight:
Ride demand is highest during daytime hours, especially when people are active.

🔹 3. Ride Distance Distribution
Most rides fall in the mid-range distance (around 3–7 km).
Very short and very long rides are comparatively fewer.

Insight:
Most users take medium-distance rides, likely for daily travel such as office or local commuting.

🔹 4. Revenue Contribution by Vehicle Type
A few vehicle types contribute major portion of revenue, while others contribute less.
Even though some vehicles have fewer rides, they still generate higher revenue per ride.

Insight:
Revenue is not only dependent on ride count but also on pricing of vehicle types.
