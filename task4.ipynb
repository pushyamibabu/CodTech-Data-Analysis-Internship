```python
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt

# Updated text data (clear positive & negative)
data = [
    "Driver was very polite",
    "Ride was very bad and frustrating",
    "Excellent service",
    "Driver was rude and cancelled ride",
    "Very smooth experience"
]

# Sentiment analysis
sentiments = []

for text in data:
    polarity = TextBlob(text).sentiment.polarity
    
    if polarity > 0:
        sentiments.append("Positive")
    elif polarity < 0:
        sentiments.append("Negative")
    

# Create DataFrame
df_sent = pd.DataFrame({
    "Text": data,
    "Sentiment": sentiments
})

# Print to verify (optional but useful)
print(df_sent)

# Plot graph
df_sent["Sentiment"].value_counts().plot(
    kind="bar",
    color=["green", "red"]
)

plt.title("Sentiment Analysis Results")
plt.xlabel("Sentiment")
plt.ylabel("Count")

plt.show()
```

INSIGHT:
The sentiment analysis shows a mix of positive and negative feedback. Positive sentiments indicate good service experiences, while negative sentiments highlight issues such as delays or cancellations. This helps in identifying areas where service can be improved
