```python
import pandas as pd

data = {
    "review": [
        "This ride was amazing",
        "Very bad experience",
        "Driver was okay",
        "Excellent service",
        "Not good at all"
    ]
}

df = pd.DataFrame(data)
df
```

```python
from textblob import TextBlob

df["sentiment"] = df["review"].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
def label(x):
    if x > 0:
        return "Positive"
    elif x < 0:
        return "Negative"
    else:
        return "Neutral"

df["label"] = df["sentiment"].apply(label)

df
```

```python
df["label"].value_counts().plot(kind='bar')
```

INSIGHTS:

1. Sentiment analysis was performed on textual data.
2. Reviews were classified as positive, negative, or neutral.
3. Most reviews show varying sentiment patterns.
